```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.grid()
plt.plot(x,y)
plt.show()
```

```python
category = np.array(["wheat","rice","dal","roti"])
values = np.array([23,54,21,14])

plt.bar(category,values,color="green")

plt.title("daily food")
plt.xlabel("Food")
plt.ylabel("values")

plt.show()
```
