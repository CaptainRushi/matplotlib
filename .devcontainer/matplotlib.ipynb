```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023,2024,2025,2026])
y = np.array([59,100,36,180])

plt.plot(x,y)
plt.show()
```

```python
x = np.array([2023,2024,2025,2026])
y = np.array([59,100,36,180])
y2 = np.array([12,56,76,100])

plt.plot(x,y,marker=".",markersize="10",markerfacecolor="orange")
plt.plot(x,y2)
plt.show()
```

```python
y3 = np.array([98,34,67,89])
line_style = dict(marker=".",
                  markerfacecolor="orange",
                  markersize="10",
                  linestyle="solid",
                  linewidth='3'
                  )
plt.plot(x,y,**line_style)
plt.plot(x,y2,**line_style)
plt.plot(x,y3,**line_style)

plt.show()
```
