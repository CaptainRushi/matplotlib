```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023,2024,2025,2026])
y = np.array([59,100,36,180])
y2 = np.array([12,56,76,100])
y3 = np.array([98,34,67,89])

plt.title("Class size", fontsize="23",
          fontweight="bold")
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlabel("Years")
plt.ylabel("Students")
plt.show()
```
