```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
category = np.array(["wheat","rice","dal","roti"])
values = np.array([23,54,21,14])

plt.pie(values,labels=category,
        autopct="%1.1f%%",
        explode=[0,0,0,0.2],
        shadow=True,
        startangle=90)
plt.title("Daily Foods")

plt.show()
```
